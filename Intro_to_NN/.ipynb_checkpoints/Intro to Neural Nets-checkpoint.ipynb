```python
import numpy as np
import aidecours
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs

from aidecours import code_button as CB
aidecours.center_images(); CB()
```

A *neural network* is a simple computational graph that expresses a function (possibly complex) as a succession of simple operations.

Let's start by considering a trivial neural network that has two input neurons and an output neuron. This network represents a function 
$R_\wbf: \Rbb^2 \to \Rbb$, which can be illustrated:

<p>
<center><img src="images/reseau_simple_.png"/></center>


The network $R_\wbf$ below receive two input values: $x_1$ et $x_2$. These values are multiplied correspondingly by weights $w_1$ and $w_2$. The right-side neuron gets values $w_1 x_1$ and $w_2 x_2$, then combine it by applying the function $f$. An obtained value corresponds to the network output. 

$$R_\wbf\left({\small\vec{x_1\\[-3mm]x_2}}\right) = f(w_1 x_1+w_2 x_2)\,.$$

## Regression: Linear network

For example, if $f$ is the identity function $f(x)=x$, the network represents the operation:

$$R_\wbf\left({\small\vec{x_1\\[-3mm]x_2}}\right) = f(w_1 x_1 +  w_2 x_2) = w_1 x_1 + w_2 x_2\,.$$

So, we have here a neural network in its trivial form, which represents a scalar product between a vector $\xbf = (x_1,x_2)$ and a vector $\wbf = (w_1,w_2)$ :
$$R_\wbf({\xbf}) = \wbf\cdot \xbf\,.$$

In practice, a neural net typically has more than 2 inputs. Denote $d$ as the dimension of a network input, i.e. the number of input neurones). Thus, we have $\xbf\in\Rbb^d$ and $\wbf\in\Rbb^d$.
$$R_\wbf({\xbf}) = \sum_{i=1}^d w_i x_i = \wbf\cdot \xbf\,.$$

<center><img src="images/reseau_simple_d.png" style="height: 200px;"/></center>

### Training of a neural network
Typically, training of a neural network includes learning a parameter vector ${\bf w}$ on a training set. 


The training set is denoted as $S$ and consists $n$ observations,

$$S = \{ ({\bf x}_1, y_1), ({\bf x}_2, y_2), \ldots, ({\bf x}_n, y_n) \}\,,$$

where an *observation* is a pair $({\bf x}, y)$.

### Loss function

We have to choose a loss fucntion L\big(y',y\big)$. 
Thus, learning procedure is to solve the following problem:

$$\min_\wbf \left[\frac1n \sum_{i=1}^n L\Big(R_\wbf(\xbf_i), y_i\Big)\right].$$ 

### Quadratic loss function

Consider the example of quadratic loss function:

$$L_{\rm quad}\big(y', y\big) = (y'-y)^2\,.$$

Here, denote as $y'$ a label predicted by a neural network and as $y$ a true label of an observation $\xbf$.

```python
def quadratic_loss(y_prime, y):
    return (y_prime-y)**2

for y in (-1, 2, 3):
    aidecours.show_1d_function(lambda a: quadratic_loss(a, y), -2, 4, label=f'$y={y}$');

plt.xlim(-2,4), plt.ylim(0,5)
plt.title('Quadratic loss'); plt.xlabel('$y\'$'); plt.ylabel('$L_{quad}(y,y\')$'); CB()
```

Thus, we have the following learning problem
$$\min_\wbf \left[\frac1n \sum_{i=1}^n (\wbf\cdot\xbf_i- y_i)^2\right].$$ 

Imagine we have a training set of 3 observations:

$$
S =  \Big\{\, \big([1,1], {-}1\big), \,
\big([0,-1], 3\big), \,
\big([2,0.5], 2\big)\,\Big\}
$$

Then, the objective function can be illustrated as follows:

```python
x = np.array([(1,1),(0,-1),(2,.5)])
y = np.array([-1,3,2])
def calc_quadratic_loss(w): return np.mean((x @ w - y) ** 2)

aidecours.show_2d_function(calc_quadratic_loss, -10, 10, .5)
plt.title('Objective function'); plt.xlabel('$w_1$'); plt.ylabel('$w_2$'); CB()
```

### This is the least squares method!

Fot this problem we know how to find the exact solution.

By representing observations $S$ in the matrix form,

$$ 
\Xbf = \left[ \begin{array}{cc}
1& 1\\
0& -1\\
2& \frac12
\end{array} \right],
\quad\ybf = \left[ \begin{array}{c}
-1\\
3\\
2
\end{array} \right]
$$

We know that
$\wbf^* = (\Xbf^T \Xbf)^{-1} \Xbf^T \ybf \approx (1.76, -2.90)$

see:https://en.wikipedia.org/wiki/Least_squares

```python
w_opt = np.linalg.inv(x.T @ x) @ x.T @ y
aidecours.show_2d_function(calc_quadratic_loss, -10, 10, .5)
plt.scatter(*w_opt, s=200, marker='*', c='r')
plt.title('Objective function'); plt.xlabel('$w_1$'); plt.ylabel('$w_2$'); CB()
```

### Regularization 

Like all learning algorithms, a neural net can *overfit* training data. We will see a few techniques that can prevent a neural net from overfitting.

### Regularization $\mathrm{L}^2$

The one way is to add a regularization term to the objective function. In many cases squared Euclidean norm of parameters $\wbf$ is used as the regularization funtion:

$$\min_\wbf \left[\frac1n \sum_{i=1}^n L\big(R_\wbf(\xbf_i), y_i\big) + \frac\lambda2\|\wbf\|^2\right].$$ 
Here, $\lambda>0$ is a *hyperparameter* of a neural network learning.

```python
lambdas = (0., 1., 10.)
fig, axes = plt.subplots(1, len(lambdas), figsize=(16, 4))
for param_lambda, ax in zip(lambdas, axes):
    def calc_regularized_quadratic_loss(w): 
        return np.mean((x @ w - y) ** 2) + param_lambda * (w @ w) / 2
    aidecours.show_2d_function(calc_regularized_quadratic_loss, -10, 10, .5, ax=ax) 
    w_opt = np.linalg.inv(x.T @ x + param_lambda * np.eye(2) * 2) @ x.T @ y
    ax.scatter(*w_opt, s=200, marker='*', c='r')
    plt.title(f'Objective function ($\lambda = {param_lambda}$)'); plt.xlabel('$w_1$'); plt.ylabel('$w_2$');
CB()
```

### It is Ridge regression problem (regularized Least Squares)

We know how to find an exact solution of the optimization problem.

By representing training data $S$ in the matrix form,

$$ 
\Xbf = \left[ \begin{array}{cc}
1& 1\\
0& -1\\
2& \frac12
\end{array} \right],
\quad\ybf = \left[ \begin{array}{c}
-1\\
3\\
2
\end{array} \right],
$$
we know that
$\wbf^* = (\Xbf^T \Xbf+\lambda n\Ibf )^{-1} \Xbf^T \ybf\,.$

$\lambda$ |$\longrightarrow$  | $w_1^*$ | $w_2^*$ 
----------|-|---------|--------
0         |$\longrightarrow$  | 1.76 | -2.90 
1         |$\longrightarrow$   | 0.73 | -1.05 
10        |$\longrightarrow$   | 0.13 | -0.15 


see: https://en.wikipedia.org/wiki/Tikhonov_regularization

## Classification: Sigmoid neuron

Combination a linear output neuron and quadratic loss function leads to regression problem, where labels are real numbers $y\in\Rbb$.

Now, let's consider a binary classification problem:

$$y\in\{0,1\}$$

Let's look again at the trivial network we considered before.

<p>
<center>
<img src="images/reseau_simple_.png"/>
</center>

For an output neuron we can appply nonlinear function $f$ called **sigmoid function** and denoted as $\sigma$: 

$$f(x) = \sigma(x) = \frac{1}{1+e^{-x}}\,.$$

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
   
aidecours.show_1d_function(sigmoid, -6, 6, constante_x=.5); 
plt.title('Sigmoid function'); plt.xlabel('$x$'); plt.ylabel('$\sigma(x)$'); CB()
```

The output of a neural network will be bounded between $0$ and $1$.

We say that the label $y=0$ is predicted when

$$R_\wbf(\xbf) = \sigma(\wbf\cdot\xbf) < 0.5\,,$$
and $y=1$ otherwise.

An output value close to $0.5$ refers to an uncertainty in the obtained result. When the output value close to $0$ or $1$, we can say that a network is *confident* in its results.

There is *bayesian* interpretation of this problem formulation, where the sigmoidal output is considered as the probability of the obtaining a label $y=1$ by a neural network $R_\wbf$ given the input $\xbf$:

$$P(y = 1\,|\,\xbf; \wbf) \ = \ \sigma(\wbf\cdot\xbf) \,.$$
And:
$$P(y = 0\,|\,\xbf; \wbf)  \ = \ 1 - P(y = 1\,|\,\xbf; \wbf) \ =\ 1 - \sigma(\wbf\cdot\xbf) \,.$$

During learnign, we would like to penalize the network when the predicted result is far from the true label. The used loss function in this case is called **negative log likelihood**:
 
$$\begin{align}
L_{\rm nlv}\big(y', y\big) = &- y \log(y') - (1-y)\log(1-y') \\[2mm]
= &\begin{cases}
-\log(1-y') & \mbox{si $y=0$ ,} \\[1mm]
-\log(y') & \mbox{si $y=1$  .}
\end{cases}\end{align}$$


```python
def loss_nlv_y0(y_prime):
    return - np.log(1. - y_prime) 
def loss_nlv_y1(y_prime):
    return - np.log(y_prime) 

aidecours.show_1d_function(loss_nlv_y0, 0, 0.999, .01, label='$y=0$');
aidecours.show_1d_function(loss_nlv_y1, 0.001, 1, .01, label='$y=1$');
plt.title('negative log likelihood'); plt.xlabel('$y\'$'); plt.ylabel('$L_{nlv}(y,y\')$'); CB()
```

#### Recap.

1. At the output neuron the sigmoid function is applied to the sum of its inputs:

$$\sigma(\wbf\cdot\xbf) = \frac{1}{1+e^{-\wbf\cdot\xbf}}\,.$$
2. If $\sigma(\wbf\cdot\xbf) > 0.5$, we say $y=1$.

3. During learning, we penalize according 

$$
\begin{align}
L_{\rm nlv}\Big(\sigma(\wbf\cdot\xbf), y\Big) 
&= - y \log(\sigma(\wbf\cdot\xbf)) - (1-y)\log(1-\sigma(\wbf\cdot\xbf))\\
&= \quad\vdots\\[1mm]
&= - y \wbf\cdot\xbf + \log(1+e^{\wbf\cdot\xbf})
\end{align}$$

The objective function is equal to:
$$
\min_\wbf \left[\frac1n \sum_{i=1}^n - y_i \wbf\cdot\xbf_i + \log(1+e^{\wbf\cdot\xbf_i})+ \frac\lambda2\|\wbf\|^2\right].
$$ 

### We end up with logistic regression problem!

```python
aidecours.show_1d_function(lambda a: loss_nlv_y0(sigmoid(a)), label='$y=0$');
aidecours.show_1d_function(lambda a: loss_nlv_y1(sigmoid(a)), label='$y=1$');
plt.title('Logistic loss'); plt.xlabel('$x$'); plt.ylabel('$L_{nlv}(1,\sigma(x))$'); CB()
```

### Example of classification problem

Below, we generate randomly a training set of 100 observations in two dimensions by using the function
`sklearn.datasets.make_blobs` 

(voir: http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html)

```python
x, y = make_blobs(n_samples=100, centers=2, n_features=2, cluster_std=1, random_state=0)

aidecours.show_2d_dataset(x, y); CB()
```

Let's illustrate the function we would like to optimize (with $\lambda=0.01$):
    
$$
\frac1n \sum_{i=1}^n - y_i \wbf\cdot\xbf_i + \log(1+e^{\wbf\cdot\xbf_i})+ \frac\lambda2\|\wbf\|^2\,.
$$ 

```python
def calc_logistic_loss(w):
    pred = sigmoid(x @ w)
    pred[y==0] = 1-pred[y==0]
    return np.mean(-np.log(pred)) + .01*w @ w/2
    
aidecours.show_2d_function(calc_logistic_loss, -4, 4, .05)
w_opt =[ 1.59526315, -1.09703947]
plt.scatter(*w_opt, s=200, marker='*', c='r')
plt.title('Objective function'); plt.xlabel('$w_1$'); plt.ylabel('$w_2$'); CB()    
```

The optimum is $\wbf^* \approx \vec{1.60\\ -1.10}\,$, which correspond to the following predictor.

```python
from sklearn.linear_model import LogisticRegression
algo = LogisticRegression(C=100./len(y), fit_intercept=False)
algo.fit(x, y)
coord=aidecours.show_2d_predictions(x, y, algo.predict)
plt.plot([coord[0],0,0], [0,0,coord[2]], '--k');
#print('vecteur w :', algo.coef_[0])
#print('biais b   :', algo.intercept_)
CB()
```

## Adding bias

To avoid the constraint of pasing the function through the origin, we add a **bias** term:

<center><img src="images/reseau_simple_biais.png" style="height: 200px;"/></center>

In this case of the trivial network, the prediction function is:

$$R_{\wbf,b}({\xbf}) = f\left(\sum_{i=1}^d w_i x_i + b\right) = f\big(\wbf\cdot \xbf+b\big).$$

Let's go back to our example of the logistic regression, i.e. the neural network where the output neuron is the **sigmoid function** $\sigma$ and the loss is **negative log likelihood** $L_{\rm nlv}$. The optimization problem with **bias** term can be expressed as follows:

$$
\min_{\wbf,b} \left[\frac1n \sum_{i=1}^n - y_i (\wbf\cdot\xbf_i+b) + \log(1+e^{\wbf\cdot\xbf_i+b})+ \frac\lambda2\|\wbf\|^2\right].
$$ 

```python
algo = LogisticRegression(C=100/len(y), fit_intercept=True)
algo.fit(x, y)
coord=aidecours.show_2d_predictions(x, y, algo.predict)
plt.plot([coord[0],0,0], [0,0,coord[2]], '--k'); CB()
#print('vecteur w :', algo.coef_[0])
#print('biais b   :', algo.intercept_[0])
CB()
```

The optimum values are $\wbf^* \approx \vec{0.97\\ -1.52}, b \approx 2.33\,$.

## Choice of the data representation. 
Let's transform the data by applying the function $\Phibf:\Rbb^2\to\Rbb^4$:

$${\Phibf} : \vec{x_1 \\ x_2} \, \mapsto \, \vec{x_1 \\ x_2 \\ x_1^2\\ x_2^2}$$



<center><img src="images/reseau_simple_phi.png"/></center>

By applying the logistic regression to the transformed data, we get:

$$
\min_{\wbf,b} \left[\frac1n \sum_{i=1}^n - y_i (\wbf\cdot\Phibf(\xbf_i)+b) + \log(1+e^{\wbf\cdot\Phibf(\xbf_i)+b})+ \frac\lambda2\|\wbf\|^2\right].
$$ 

```python
def phi(x):
    return np.hstack((x, (x[:,0]**2).reshape(-1,1), (x[:,1]**2).reshape(-1,1)))

algo = LogisticRegression()
algo.fit(phi(x), y)
aidecours.show_2d_predictions(x, y, lambda a: algo.predict(phi(a))) 
CB()
```

## Neural network with one hidden layers

By adding a **hidden layer** to the network, we allow learning of appropriate data transformation.

<p>
<center><img src="images/reseau_simple_2couches.png"/></center>

The network $R_{\Vbf,\wbf}(\xbf) = f\Big(\wbf\, h\big(\Vbf \xbf\big)\Big)$ is parametrized by the matrix $\Vbf$ and the vector $\wbf$:

$$\xbf = \vec{x_1\\x_2} ; \quad
\Vbf=\vec{v_{11} & v_{12} \\
          v_{21} & v_{22} \\ 
          \vdots & \vdots \\
          v_{D1} & v_{D2} }; \quad
\wbf = \vec{w_1 & w_2 & \cdots& w_D}
$$



## Activation function
We've already seen two possible choices of the *activation function* for the network output:
* Linear function: $f(a) = a$
* Sigmoid function $f(a) = \sigma(a) = \dfrac{1}{1+e^{-a}}\,.$

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
   
aidecours.show_1d_function(sigmoid, -6, 6, constante_x=0, linewidth=5); 
plt.title('Fonction sigmoidale'); plt.xlabel('$a$'); plt.ylabel('$\sigma(a)$'); CB()
```

## Activation function

<center><img src="images/reseau_simple_2couches.png"/></center>

We will examine two other possible choices of the function $h$ for a hidden layer:
* Hyperbolic tangent (tanh)
* *Rectified linear unit* (ReLU)

### Hyperbolic tangent

$$h(x) = \tanh(x) = 2\sigma\big(2x\big) - 1$$

```python
aidecours.show_1d_function(np.tanh, -6, 6, constante_x=0, linewidth=5, label='$\\tanh(x)$'); 
aidecours.show_1d_function(lambda x: sigmoid(x), -6, 6, c='b', linestyle='-.', linewidth=2, label='$\sigma(x)$'); 
plt.title('Function of hyperbolic tangent'); plt.xlabel('$x$'); plt.ylabel('$\mathrm{tanh}(x)$'); CB()
```

### Rectified Linear Unit (ReLU)
$$h(x) = \mathrm{relu}(x) = \max(0, x)$$

```python
def relu(x):
    return np.maximum(0, x)
aidecours.show_1d_function(relu, -6, 6, constante_x=0, linewidth=5); 
plt.title('«relu» function'); plt.xlabel('$x$'); plt.ylabel('$\mathrm{relu}(x)$'); CB()
```
